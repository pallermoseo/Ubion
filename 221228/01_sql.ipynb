# pandas 복습
1. 새로운 파생변수 생성 job_code
    - salesman은 1, manager는 2, 그외의 직업은 3
    - 컬럼은 EMPNO, ENAME, job_code만 출력

```python
# apply, def
#def change(x):
    #if x == 'SALESMAN':
    #    return 1
    #elif x == 'MANAGER':
    #    return 2
    #else:
    #    return 3

#emp['JOB'].apply(change)

# apply, lambda
# emp['JOB'].apply(lambda x : 1 if (x == 'SALESMAN') else 2 if (x == 'MANAGER') else 3))

# for 문
emp2 = emp.copy()

for i in emp2.index:
    if emp2.loc[i, 'JOB'] == 'SALESMAN':
        emp2.loc[i, 'job_code'] = 1
    elif emp2.loc[i, 'JOB'] == 'MANAGER':
        emp2.loc[i, 'job_code'] = 2
    else:
        emp2.loc[i, 'job_code'] = 3
emp2[['EMPNO', 'ENAME', 'job_code']]
    
```

```python
# 위의 결과를 sql 쿼리문을 이용하여 생성
sql = ''' 
    select
    EMPNO,
    ENAME,
    JOB,
    CASE
        when JOB = 'SALESMAN'
        then 1
        when JOB = 'MANAGER'
        then 2
        else 3
    END as job_code 
    from
    EMP
'''
sql_class.executeAll(sql)
```

# pandas
1. tran_1, tran_2, tran_d_1, tran_d_2 데이터베이스에서 select문을 이용해서 데이터프레임으로 저장
2. tran_1, tran_2 단순한 행 결합(유니언 결합)
3. tran_d_1, tran_d_2 단순한 행 결합(유니언 결합)
4. tran, tran_d를 특정 조건에 맞게 조인 결합

```python
sql = ''' 
    select
    *
    from
    tran_1
'''
tran_1 = sql_class.executeAll(sql)
```

```python
sql = ''' 
    select
    *
    from
    tran_2
'''
tran_2 = sql_class.executeAll(sql)
```

```python
sql = ''' 
    select
    *
    from
    tran_d_1
'''
tran_d_1 = sql_class.executeAll(sql)
```

```python
sql = ''' 
    select
    *
    from
    tran_d_2
'''
tran_d_2 = sql_class.executeAll(sql)
```

```python
tran = pd.concat([tran_1, tran_2], ignore_index=True)
tran.head(2)
```

```python
tran_d = pd.concat([tran_d_1, tran_d_2], ignore_index=True)
tran_d.head(2)
```

```python
transaction = pd.merge(tran, tran_d, on = 'transaction_id', how= 'left')
transaction.head(1)
```

```python
# sql 쿼리문
sql = ''' 
    select * from tran_1
    union
    select * from tran_2
'''
sql_class.executeAll(sql)
```

```python
sql = ''' 
    select * from 
    (select * from tran_1 
    union select * from tran_2) as tran
    left join
    (select * from tran_d_1
    union select * from tran_d_2) as tran_d
    on
    tran.transaction_id = tran_d.transaction_id
'''
sql_class.executeAll(sql)
```
