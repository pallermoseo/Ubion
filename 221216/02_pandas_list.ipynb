```python
import pandas as pd
```

```python
# 파일의 목록 출력
import os
```

```python
file_list = os.listdir('../csv/2017/')
```

```python
# 파일 리스트는 출력 성공
# 파일이 12개가 존재
# 12개의 파일을 하나의 데이터프레임에 결합

#'../csv/2017/'+file_name

file_path = '../csv/2017/'
pd.read_csv(file_path+file_list[0])
# 비어있는 데이터프레임 생성 
df = pd.DataFrame()
```

```python
# 파일리스트, path 지정해놓은 변수
# 반복문 사용
# 어떠한 값을 가지고 반복해야 하는가? -> 파일 리스트의 값들을 반복
# concat() 사용할 데이터프레임을 로드
# 비어있는 데이터프레임에 로드한 데이터프레임을 결합

for i in file_list:
    data = pd.read_csv(file_path+i)
    df = pd.concat([df, data], axis="rows", ignore_index=True)



#for i in range(0, len(file_list),1):
    #data = pd.read_csv(file_path+file_list[1])
    #df = pd.concat([df, data], axis=0, ignore_index=True)
    

    
```
