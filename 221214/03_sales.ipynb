```python
import pandas as pd
```

#csv파일 로드 데이터를 튜닝
1. csv 폴더에 있는 sales records.csv 파일 로드
2. 'Order ID','Total Revenue','Total Cost','Total Profit' 컬럼을 삭제
3. 컬럼의 이름 변경['권역','국가','상품종류','채널','우선순위','주문일자','발송일자','판매단위','단가','원가']
4. 권역, 국가를 기준으로 오름차순 정렬
5. index의 값을 초기화하고 기존의 인덱스는 삭제

```python
sales=pd.read_csv('../csv/Sales Records.csv')
sales
```

```python
sales.drop(['Order ID','Total Revenue','Total Cost','Total Profit'], axis=1, inplace=True)
sales
```

```python
sales.columns=['권역','국가','상품종류','채널','우선순위','주문일자','발송일자','판매단위','단가','원가']
sales
```

```python
sales.sort_values(['권역','국가'], inplace=True)
sales
```

```python
sales.reset_index(drop=True, inplace=True)
sales
```

```python
#단가 컬럼의 float 형태의 데이터를 반올림 -> int의 형태로 변경
#반올림 함수 round(n)
#n의 값은 소수점의 자릿수 -> n은 최종 숫자에 남는 갯수 (n이 1이면 소수점 한 자리까지 남음)
#sales['단가'].round(n) -> -1의 경우 1의 자리까지 숫자가 없어지고 반올림됨
#데이터의 형태는 스리즈 values 형태는 float
#데이터 타입을 변경 함수
#astype('int')

#sales['단가'].round(0).astype('int') -> 두 작업을 동시 수행
sales['단가']=sales['단가'].round(0).astype('int') 
```

```python
#원가 철럼의 데이터를 반올림 하지 않고 int 타입으로 변경
sales['원가'].astype('int')
```

```python
sales['원가']=sales['원가'].astype('int')
```

```python
sales.head()
```

1. 새로운 파생변수 (총 단가, 총 원가, 총 이윤) 생성
2. 총 단가 = 단가 * 판매단위
3. 총 원가 = 원가 * 판매단위
4. 총 이윤 = 총 단가 - 총 원가

```python
# sales['총 단가']
sales['총 단가']=sales['단가']*sales['판매단위']
sales['총 원가']=sales['원가']*sales['판매단위']
```

```python
sales['총 이윤']=sales['총 단가']-sales['총 원가']

sales.head()
```

```python
#loc[]
#특정 행이나 특정 열을 추출할때 사용
#loc[행의 위치, 열의 위치]

#sales 데이터에서 인덱스가 11부터 마지막까지 (=인덱스를 0부터 10까지는 삭제)
sales.loc[11:]
```

```python
#해당하는 컬럼의 데이터를 그룹화하여 카운트 체크
sales['채널'].value_counts()
```

```python
#국가별로 그룹화 하여 총 이윤의 합계를 구한 다음 가장 이윤이 높은 나라 top10을 구하시오.

#1. 국가별로 그룹화
#2. 총 이윤의 합계 -> 데이터프레임
#3. 총 이윤을 토대로 내림차순 정렬
#4. 상위 10개 출력

#그룹화 함수 -> groupby(기준이 되는 컬럼)

test=sales.groupby('국가').sum()
```

```python
test[['총 이윤']]
```

```python
#총 이윤을 기준으로 내림차순 정렬
#sort_values() -> 기본값은 오름차순 정렬
#내림차순 정렬로 변경하려면 함수의 매개변수 ascending=True 오름차순(기본값), =False 내림차순
test.sort_values(['총 이윤'],ascending=False).head(10) #inplace=True or head(10) top10 보여주기 둘 다 가능
```
