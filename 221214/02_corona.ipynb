```python
import pandas as pd
```

```python
#코로나 데이터를 로드
corona_df=pd.read_csv('corona.csv')
corona_df.head()
```

```python
#데이터 프레임에서 컬럼을 삭제
#drop() 함수를 이용하여 컬럼을 삭제
#drop 함수는 axis 라는 매개변수의 값을 기준으로 행, 열을 삭제한다.
corona=corona_df.drop(['Unnamed: 0'], axis=1)

#inplace라는 매개변수 기본값이 false -> True 변경을 하면 기준이 되는 데이터의 값이 변환
#corona_df.drop(['Unnamed: 0'], axis=1, inplace=True)
#corona_df=corona_df.drop(['Unnamed: 0'], axis=1) #3가지는 컬럼을 삭제해서 보여주는 방법
```

```python
#컬럼의 이름을 변경
#데이터프레임에 있는 컬럼의 값들을 출력하는 방법
#데이터프레임명.columns
corona.columns=['등록일시','사망자','확진자','게시글번호','기준일','기준시간','수정일시','누적의심자','누적확진률']
```

```python
corona.head()
```

```python
#corona 데이터프레임에서 컬럼 '게시글번호' 를 삭제
#corona라는 변수를 수정
corona=corona.drop(['게시글번호'],axis=1)
#corona.drop(['게시글번호'], axis=1, inplace=True)
corona.head()
```

```python
#데이터프레임의 정렬을 변경
corona.sort_values(['등록일시'], inplace=True) #inplace 함수 활용, 바로 데이터를 변환
```

```python
#인덱스를 초기화하여 새로운 넘버링
corona.reset_index(drop=True, inplace=True) #reset_index 함수내 괄호안에 drop=True 넣는다.
```

```python
corona.head()
```

```python
corona.tail()
```

```python
#데이터간의 차이를 나타내는 함수
#diff()
#corona['확진자'].diff() #괄호안 숫자는 n번째 데이터와 비교할때 씀 #ex)365 -> 1년전 데이터

#corona 데이터프레임에 새로운 파생변수(일일 확진자) 추가
corona['일일 확진자']=corona['확진자'].diff() #딕셔너리의 key, value 변환과 같음.
```

```python
corona.head()
```

```python
#새로운 파생변수 일일 사망자를 추가
#누적의심자, 누적확진률 컬럼을 삭제(corona데이터프레임을 변경)

corona['일일 사망자']=corona['사망자'].diff()
corona=corona.drop(['누적의심자','누적확진률'],axis=1) #corona.drop(['누적의심자','누적확진률'],axis=1, inplace=True)와 같음
```

```python
corona.head()
```

```python
#결측치가 존재하는 행을 삭제
#dropna() 함수를 이용하여 삭제
corona.dropna(axis=0)
```

```python
#특정한 행을 삭제 - drop() 함수를 이용하여 삭제
corona.drop([0], axis=0)
```

```python
#특정한 행을 삭제
#인덱스의 값이 0부터 10까지를 삭제
corona.drop([],axis=0)
```

```python
#결측치를 다른 값으로 대체
#fillna(값) -> 결측치의 값들이 대체

#corona데이터프레임 안에 있는 모든 결측치를 0으로 대체
corona.fillna(0)

#corona데이터프레임에서 일일 확진자에 있는 결측치만 0으로 대체
corona['일일 확진자']=corona['일일 확진자'].fillna(0)
corona.head()
```
