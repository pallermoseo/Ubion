```python
kt=yf.download('030200.KS', start='2020-01-01')
kt
```

```python
# 결측치가 존재하는가?
kt.isna().sum()
```

```python
#등락폭, 등락율 파생변수 생성
#등락폭 = 종가 - 시가
#등락율 = 등락폭 / 시가 * 100

kt['등락폭'] = kt['Close']-kt['Open']
kt['등락율'] = (kt['등락폭'] / kt['Open'])*100

kt.head()
```

```python
#시각화
import matplotlib.pyplot as plt
```

```python
plt.plot(kt['등락율'].values)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(kt['Open'].values)
plt.subplot(2,2,2)
plt.plot(kt['Close'].values)
plt.subplot(2,2,3)
plt.plot(kt['Volume'].values)
plt.subplot(2,2,4)
plt.plot(kt['등락율'].values)
plt.show()
```

#그래프 4개를 출력
#subplot(행의수, 열의수, 선택할 영역) -> 영역 나누는 작업, 나눈 영역에서 영역을 선택
#ex) subplot(2,2,1) -> 2행 2열로 영역 나눈다. -> 4개로 만든다. -> 1번 영역 선택

```python
#영역 분할 및 선택
plt.subplot(4,1,1)
#그래프 그리기
plt.plot(kt['Open'].values)
#영역 선택
plt.subplot(4,1,2)
plt.plot(kt['Close'].values)
plt.subplot(4,1,3)
plt.plot(kt['Volume'].values)
plt.subplot(4,1,4)
plt.plot(kt['등락율'].values)
plt.show()
```
