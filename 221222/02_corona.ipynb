1. corona.csv 파일을 로드
2. 결측치가 존재하는지 확인
3. Unnamed: 0 컬럼 삭제
4. 컬럼의 이름 변경 -> [등록일시, 사망자, 확진자, 게시글번호, 기준일, 기준시간, 수정일시, 누적의심자, 누적확진율]
5. 등록일시 기준으로 오름차순 정렬, 인덱스 리셋 (기존의 인덱스 삭제)
6. 새로운 파생변수 일일 사망자, 일일 확진자 생성
7. 게시글번호, 기준일, 기준시간 컬럼을 삭제
8. 일일사망자, 일일확진자 안의 값이 음수인 행을 삭제

```python
import pandas as pd
```

```python
corona = pd.read_csv('../csv/corona.csv')
corona
```

```python
corona.drop(['Unnamed: 0'], axis=1, inplace=True) # corona.iloc[:, 1:] - 같은 방법
corona
```

```python
corona.columns = ['등록일시', '사망자', '확진자', '게시글번호', '기준일', '기준시간', '수정일시', '누적의심자','누적확진율']
corona
```

```python
corona.sort_values(['등록일시'], ascending=True, inplace=True) # corona.sort_values(['등록일시'], ascending=True).reset_index(drop=True, inplace=True)
corona.reset_index(drop=True, inplace=True)
corona
```

```python
corona['일일사망자'] = corona['사망자'].diff()
corona['일일확진자'] = corona['확진자'] - corona['확진자'].shift()
corona
```

```python
corona.drop(['게시글번호','기준일','기준시간'], axis=1, inplace=True)
corona
```

```python
corona.isna().sum()
```
