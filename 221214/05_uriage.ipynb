1. purchase_date 컬럼의 데이터를 시계열 데이터로 변경
2. item_name의 데이터를 모두 대문자로 변경
3. item_name의 데이터에서 공백을 삭제
4. item_name를 기준으로 오름차순 정렬 sort_values()- 내림차순 변경시 ascending=False
5. 인덱스 초기화 reset_index(drop=True)

```python
uriage['purchase_date'] = pd.to_datetime(uriage['purchase_date'])
```

```python
# 2.item_name에 있는 데이터를 모두 대문자로 통일
# upper() 함수를 이용하면 모든 문자가 대문자로 변경

uriage['item_name'] = uriage['item_name'].str.upper() #이 경우는 str 을 붙여야 한다. Series 형태이기 때문.
```

```python
# 3.item_name의 데이터에서 공백을 제거
# 공백을 제거하는 함수
# strip() -> 양쪽의 공백을 제거하는 함수 (문자열 사이의 공백은 제거하지 못한다.)
# replace() -> 지정한 문자열을 대체하는 함수 -> ' ' -> '' 대체를 하는 작업 (공백 제거)

#case1

#uriage['item_name'].str.replace(' ','')

#case2
#apply 함수

def change(x):
    result=x.replace(' ','')
    return result

# uriage['item_name'].apply(change)

#case3
#apply 함수 + lambda 함수

uriage['item_name']=uriage['item_name'].apply(lambda x : x.replace(' ',''))
```

```python
# 4. item_name 기준으로 오름차순 정렬

uriage.sort_values(['item_name'], inplace=True)
```

```python
# 5. 인덱스 초기화

uriage=uriage.reset_index(drop=True) #uriage.reset_index().drop('index', axis=1)
uriage
```

# 결측치에 데이터를 치환
1. 상품A의 item_price의 값은 결측치중 상품이 A면 이 값을 치환
2. 나머지 상품들도 마찬가지로 같은 작업을 반복

```python
flg_null=uriage['item_price'].isna()
```

```python
#결측치 값을 확인
uriage.loc[flg_null]
```

```python
#결측치가 아닌 값을 확인
#~ 연산자 : not -> True to False 변경 vice versa
uriage.loc[~flg_null]
```

```python
# 상품A인 데이터프레임 출력
uriage.loc[uriage['item_name']=='상품A']
```

```python
#상품A이고 결측치인 데이터프레임만 출력
uriage.loc[(uriage['item_name']=='상품A') & flg_null]
```

```python
uriage['item_name'].value_counts().index
```

```python
uriage['item_name'].unique()
```

```python
# 상품A인 제품들의 결측치를 제외한 item_price의 평균값 출력
uriage.loc[(uriage['item_name']=='상품A') & ~flg_null]
a_price = uriage.loc[(uriage['item_name']=='상품A') & ~flg_null, 'item_price'].mean() #mean() -> 평균구하는 함수
# a_price 변수에 상품A이면서 결측치가 아닌 데이터의 평균값을 정의
#uriage.loc[(uriage['item_name']=='상품A') & ~flg_null]['item_price'].mean()도 가능함
```

```python
#상품A인 제품들의 결측치인 데이터프레임을 출력 & 평균값으로 결측치 채우기
uriage.loc[(uriage['item_name']=='상품A')& flg_null].fillna(a_price)
```

```python
# 상품의 목록 변수에 저장
name_list = uriage['item_name'].unique()
```

```python
for i in name_list:
    price = uriage.loc[(uriage['item_name']==i) & ~flg_null, 'item_price'].mean()
    uriage.loc[(uriage['item_name']==i)& flg_null] =uriage.loc[(uriage['item_name']==i)& flg_null].fillna(price)

    # uriage.loc[flg_null & (uriage['item_name'] == i), 'item_price'] = price -> 이것도 가능

```
