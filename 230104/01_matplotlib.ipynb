```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4])
plt.show() # -> print와 마찬가지로 만들어진 최종 결과물 그래프를 출력한다는 의미
```

- subplot()
    - 한 영역에 여러개의 그래프를 그려주기 위한 함수
    - subplot(행의 개수, 열의 개수, 위치)

```python
plt.subplot(1,2,1)
plt.plot([1,2,3,4])
plt.subplot(1,2,2)
plt.plot([4,3,2,1])
plt.show()
```

```python
fig, axes = plt.subplots(2,2)
axes[0][0].plot([1,2,3,4])
axes[0][1].plot([4,3,2,1])
axes[1][0].plot([1,2,2,1])
axes[1][1].plot([3,4,2,1])
plt.show()  # subplot 일일히 다 적지 않고 쓸 수 있는 방법
```

- 그래프를 그릴때 주의할 점
    - y축의 값만 입력하는 경우 - 문제는 거의 발생하지 않는다.
    - x축과 y축의 값을 같이 입력하는 경우
        - x의 값과 y의 값의 위치를 정확하게 대입
        - x의 개수와 y의 개수가 일치

```python
x = [1,2,3,4]
y = [4,3,2,1]

plt.plot(x, y)
plt.show()

# x와 y의 개수가 다르면 그래프가 그려지지 않는다.
```

```python
# x축의 이름, y축의 이름, 그래프의 이름
plt.plot([1,2,3,4])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('title')
plt.show()
```

```python
plt.plot([1,2,3,4], label='ascending')
plt.plot([4,3,2,1], label='descending')
plt.legend(loc=5)
plt.show()
# 범례 (legend) 지정, 범례의 내용과 위치 지정
```
